```python
import numpy as np
from cvxpy import *

A = np.matrix([[2, 2], [4, 1], [2, 4]])
x = Variable(2)
c = np.array([4, 3])
b = np.array([5, 9, 9])

LP = Problem(Minimize(c.T * x),
[A * x <= b, x >= 0])

LP.solve()
```
